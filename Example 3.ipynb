```python
#5 Number is Divisible by 5 and 11
num=int(input("Enter the integer number"))
if (num%5==0) and (num%11==0):
    print(num,"is Divisible by 5 and 11")
else:
    print(num,"is not Divisible by 5 and 11" )
```
